2. Write a function that filters the dataset for questions that contains all of the words in a list of words. For example, when the list `["King", "England"]` was passed to our function, the function returned a DataFrame of 49 rows. Every row had the strings `"King"` and `"England"` somewhere in its `" Question"`.

   Test your function by printing out the column containing the question of each row of the dataset.

````python
def filter_data(data, words):
    """
    This function will find and return all rows where the question
    contains the strings "King" and "England" exactly as they
    were passed to the function. It will not return rows where
    "King" and "England" appear as a substring in other words since
    we are using the all() keyword.
    """
    filter = lambda x: all(word in x for word in words)
    return data.loc[data["question"].apply(filter)]

filtered = filter_data(data, ["King", "England"])
filtered.head(8)
````

5. Write a function that returns the count of unique answers to all of the questions in a dataset. For example, after filtering the entire dataset to only questions containing the word `"King"`, we could then find all of the unique answers to those questions. The answer "Henry VIII" appeared 55 times and was the most common answer.

````python
def unique_answers(df, word_filter):
    question_filter = r'\b(?:{})\b'.format('|'.join(map(re.escape, word_filter)))    
    temp_df = df[df['question'].str.contains(question_filter)]
    return temp_df, temp_df['answer'].nunique()

a, b = unique_answers(data, ['march', 'bird', 'ten'])
print(f'{a}\nFrom this dataset, there are {b} unique answers.')
````
