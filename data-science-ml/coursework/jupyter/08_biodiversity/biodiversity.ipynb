# Biodiversity Portfolio Project

This project will explore and visualize data from two workbooks: `observations.csv` and `species_info.csv`. These two workbooks contain information regarding the observations of different species in different national parks, the conservation status of thse species, and other information like common name.

## Imports and Loading the Data

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

sns.set_palette('pastel')
```

```python
ob_df = pd.read_csv('observations.csv')
sp_df = pd.read_csv('species_info.csv')
```

## Exploration

Starting with the observation dataframe, a nice thing to know may be the number of species observed.

```python
print(f'Number of unique species: {sp_df.scientific_name.nunique()}')
```

```python
print(f"Number of categories of species: {sp_df.category.nunique()}")
print(f"Unique categories of species: {sp_df.category.unique()}")
```

```python
sns.countplot(data=sp_df, x='category', hue='category')
plt.title('Count of Unique Categories of Observed Species')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
plt.clf()
```

## Protected Species

```python
filtered_df = sp_df.dropna()
order = ['In Recovery', 'Threatened','Species of Concern','Endangered']
sns.countplot(data=filtered_df, 
              x='conservation_status', 
              hue='conservation_status', 
              order=order)
plt.title('Count of Conservation Statuses')
plt.xlabel('Status')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
plt.clf()
```

```python
filtered_sp = ob_df.dropna()
sns.barplot(x=filtered_sp['park_name'], 
              y=filtered_sp['scientific_name'].nunique())
plt.title('Count of Unique Species per Park')
plt.xlabel('Park')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
plt.clf()
```
