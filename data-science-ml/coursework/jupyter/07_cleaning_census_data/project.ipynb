## Regex to the Rescue

5. Use regex to turn the `Income` column into a format that is ready for conversion into a numerical type.

````python
df.Income = df["Income"].replace(r"[\$,]", "", regex=True)
df.Income = pd.to_numeric(df.Income)
df.head()
````

6. Look at the `GenderPop` column. We are going to want to separate this into two columns, the `Men` column, and the `Women` column.

   Split the column into those two new columns using `str.split` and separating out those results.

````python
gender_split = df["GenderPop"].str.split(r"(\d+)", expand=True)
df["male_count"] = pd.to_numeric(gender_split[1])
df["female_count"] = pd.to_numeric(gender_split[3])
print(df.head())
````

7. Convert both of the columns into numerical datatypes.

   There is still an `M` or an `F` character in each entry! We should remove those before we convert.

8. Now you should have the columns you need to make the graph and make sure your boss does not slam a ruler angrily on your desk because you've wasted your whole day cleaning your data with no results to show!

````python
plt.scatter(df.male_count, df.Income, alpha=0.5)
plt.scatter(df.female_count, df.Income, alpha=0.5)
plt.show()
plt.clf()
````

9. You want to double check your work. You know from experience that these monstrous csv files probably have `nan` values in them! Print out your column with the number of women per state to see.

   We can fill in those `nan`s by using pandas' `.fillna()` function.
   
   You have the `TotalPop` per state, and you have the `Men` per state. As an estimate for the `nan` values in the `Women` column, you could use the `TotalPop` of that state minus the `Men` for that state.
   
   Print out the `Women` column after filling the `nan` values to see if it worked!

````python
print(df.female_count.isna().sum())
df["female_count"] = df["female_count"].fillna(df["TotalPop"]-df["male_count"])
print(df["female_count"])
````

## Histogram of Races

13. Now your boss wants you to make a bunch of histograms out of the race data that you have. Look at the `.columns` again to see what the race categories are.

````python
# dr. claude helped write this
# this will replace all columns
# using iloc accessors from cols
# 3 to 8 (upperbound) by using a
# vectorized lambda apply function
# to replace % with nothing and then
# convert it to a decimal

df.iloc[:, 3:9] = df.iloc[:, 3:9].apply(lambda x: pd.to_numeric(x.str.replace('%', '')) / 100)

df.head()
````

````python
# count the number of nan in each
# race percent column
df.iloc[:, 3:9].isna().sum()
````

````python
subset_cols = df.columns[3:8]  # columns 3 through 8 (zero-based indexing)
row_sum = df.iloc[9][subset_cols].sum(skipna=True)
df.iloc[9] = df.iloc[9].fillna(1 - row_sum)

# If you want to apply this to all rows:
def fill_remaining(row):
    subset_sum = row[subset_cols].sum(skipna=True)
    return row.fillna(1 - subset_sum)

df = df.apply(fill_remaining, axis=1)

# count the number of nan in each
# race percent column
df.iloc[:, 3:9].isna().sum()
````

14. Try to make a histogram for each one!

    You will have to get the columns into the numerical format, and those percentage signs will have to go.
    
    Don't forget to fill the `nan` values with something that makes sense! You probably dropped the duplicate rows when making your last graph, but it couldn't hurt to check for duplicates again.

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Approach 1: 6 rows x 9 columns (54 total spots, with some empty)
fig, axes = plt.subplots(6, 9, figsize=(25, 17))
axes = axes.ravel()

ethnic_columns = ['Hispanic', 'White', 'Black', 'Native', 'Asian', 'Pacific']
colors = sns.color_palette("husl", len(ethnic_columns))

# Plot each state/territory
for idx, row in enumerate(df.itertuples()):
    if idx < 51:  # Now including all 51 regions
        values = [getattr(row, col) * 100 for col in ethnic_columns]
        
        axes[idx].bar(range(len(ethnic_columns)), values, color=colors)
        axes[idx].set_title(row.State, fontsize=8)
        axes[idx].tick_params(axis='both', which='major', labelsize=6)
        axes[idx].set_xticks(range(len(ethnic_columns)))
        axes[idx].set_xticklabels(ethnic_columns, rotation=45, ha='right')
        axes[idx].set_ylim(0, 100)
        
        # Only show y-axis for plots on the left edge
        if idx % 9 != 0:
            axes[idx].set_yticks([])

# Hide empty subplots
for idx in range(51, 54):
    axes[idx].set_visible(False)

plt.suptitle('Ethnic Distribution in U.S. States and Territories', fontsize=16, y=1.02)


legend_elements = [plt.Rectangle((0,0),1,1, color=color) for color in colors]
fig.legend(legend_elements, ethnic_columns, 
          loc='center left', 
          bbox_to_anchor=(1, 0.5),
          title='Ethnicities')

plt.tight_layout()
plt.show()
````
